```
import inkml2img, os, glob
```

```
import lxml.etree as xET
```

```
parser = xET.XMLParser(recover=True)
```

```
!mkdir images
```

```
!mkdir truth
```

```
for filename in os.listdir('./data'):
  if filename.endswith('.inkml'):
    tree = xET.parse('./data/' + filename, parser=parser)
    root = tree.getroot()
    
    wrapper_xml = root.find('{http://www.w3.org/2003/InkML}annotationXML')
    math_xml = wrapper_xml.find('{http://www.w3.org/1998/Math/MathML}math')
    
    xET.strip_attributes(root, '{http://www.w3.org/XML/1998/namespace}id')
    
    mathml_str = xET.tostring(math_xml, encoding="unicode")
#     print(filename + " " + " ".join(mathml_str.split()))
    print(filename)
    filename_without_ext = os.path.splitext(filename)[0]
    file = open('./truth/' + filename_without_ext + '.txt', 'w')
    file.write(mathml_str) 
    file.close()
    
    inkml2img.inkml2img('./data/' + filename,'./images/' + os.path.splitext(filename)[0] + '.png')
```
